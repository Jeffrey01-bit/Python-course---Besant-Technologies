```python
import pandas as pd

# --------------------------------------------------
# EMPLOYEE DATAFRAME
# --------------------------------------------------

employee_df = pd.DataFrame({
    'Emp_ID': [101,102,103,104,105,106,107,108],
    'Emp_Name': ['John','Alice','Bob','David',
                 'Eva','Frank','Grace','Helen'],
    'Dept_ID': [1,2,1,3,2,4,5,None],
    'Salary': [60000,55000,70000,50000,
               65000,72000,48000,53000]
})

print("EMPLOYEE DATAFRAME")
print(employee_df)

# --------------------------------------------------
# DEPARTMENT DATAFRAME
# --------------------------------------------------

department_df = pd.DataFrame({
    'Dept_ID': [1,2,3,4,6],
    'Department': ['IT','HR','Finance','Sales','Admin'],
    'Location': ['Chennai','Mumbai','Delhi','Pune','Bangalore']
})

print("\nDEPARTMENT DATAFRAME")
print(department_df)
```

```python
# ==================================================
# 1. INNER JOIN
# ==================================================

inner_join = pd.merge(
    employee_df,
    department_df,
    on='Dept_ID',
    how='inner'
)

print("\nINNER JOIN")
print(inner_join)
```

```python
# ==================================================
# 1. Left JOIN
# ==================================================

left_join = pd.merge(
    employee_df,
    department_df,
    on='Dept_ID',
    how='left'
)

print("\nLEFT JOIN")
print(left_join)
```

```python
# ==================================================
# 3. RIGHT JOIN
# ==================================================

right_join = pd.merge(
    employee_df,
    department_df,
    on='Dept_ID',
    how='right'
)

print("\nRIGHT JOIN")
print(right_join)
```

```python
# ==================================================
# 4. FULL OUTER JOIN
# ==================================================

outer_join = pd.merge(
    employee_df,
    department_df,
    on='Dept_ID',
    how='outer'
)

print("\nFULL OUTER JOIN")
print(outer_join)
```

```python
# ==================================================
# 5. CROSS JOIN
# ==================================================

cross_join = pd.merge(
    employee_df,
    department_df,
    how='cross'
)

print("\nCROSS JOIN")
print(cross_join)  # Showing first 5 rows only
```

```python
import pandas as pd

# --------------------------------------------------
# Table 1 : Employee Table
# --------------------------------------------------
employees = pd.DataFrame({
    'Emp_ID': [101,102,103,104,105,106,107,108],
    'Emp_Name': ['John','David','Sara','Priya','Arun','Smith','Kiran','Meena'],
    'Dept_ID': [1,2,1,3,2,1,3,2],
    'Location_ID': [11,12,11,13,12,11,13,12],
    'Salary': [65000,72000,50000,85000,67000,62000,91000,70000]
})

print("EMPLOYEE TABLE")
print(employees)


# --------------------------------------------------
# Table 2 : Department Table
# --------------------------------------------------
departments = pd.DataFrame({
    'Dept_ID': [1,2,3],
    'Department': ['IT','HR','Finance']
})

print("\nDEPARTMENT TABLE")
print(departments)


# --------------------------------------------------
# Table 3 : Location Table
# --------------------------------------------------
locations = pd.DataFrame({
    'Location_ID': [11,12,13],
    'Location': ['Chennai','Mumbai','Delhi']
})

print("\nLOCATION TABLE")
print(locations)


# ==================================================
# JOIN 1 :
# Employee + Department
# ==================================================
emp_dept = pd.merge(
    employees,
    departments,
    on='Dept_ID',
    how='inner'
)

print("\nEMPLOYEE + DEPARTMENT")
print(emp_dept)


# ==================================================
# JOIN 2 :
# Previous result + Location
# ==================================================
final_df = pd.merge(
    emp_dept,
    locations,
    on='Location_ID',
    how='inner'
)

print("\nFINAL JOINED TABLE")
print(final_df)


# ==================================================
# GROUP BY with MULTIPLE AGGREGATIONS
# ==================================================
result = final_df.groupby(
    ['Department', 'Location']
).agg(
    Employee_Count=('Emp_ID', 'count'),
    Avg_Salary=('Salary', 'mean'),
    Max_Salary=('Salary', 'max'),
    Total_Salary=('Salary', 'sum')
).reset_index()

print("\nGROUP BY RESULT")
print(result)
```
