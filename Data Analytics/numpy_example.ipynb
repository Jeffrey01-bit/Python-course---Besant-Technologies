```python
# Single dimension
import numpy as np
# total for five students
total = []
for i in range(0, 5):
    a = int(input(f"Student {i+1}, Total = "))
    total.append(a)
output = np.array(total)
print(total)
print(output)
```

```python
#Two dimension
#average of 5 students marks in 5 subjects
import numpy as np
li=[]
avg=[]
for i in range(0,5): #5 students
    li1=[]
    sum=0
    for j in range(0,5): #5 subject marks
        a=int(input(f"Student{i+1}, Mark in sub{j+1}"))
        li1.append(a)
        sum+=a
    li.append(li1)
    avg.append(sum/5)
print("List representation:\n",li)
arr=np.array(li)
print("Array representation: \n",arr)
print(avg)
```

```python
#Three dimension
import numpy as np
no_students= int(input("Enter number of students: "))
no_semesters= int(input("Enter number of semesters: "))
no_subjects= int(input("Enter number of subjects: "))
p_list=[]
for i in range(0,no_students):
    sem_list=[]
    for j in range(0, no_semesters):
        marks_list=[]
        for k in range(0, no_subjects):
            m=int(input(f"Student: {i+1} Semester: {j+1}, Enter mark in subject {k+1}: "))
            marks_list.append(m)
        sem_list.append(marks_list)
    p_list.append(sem_list)
print(p_list)
print(np.array(p_list))
```

```python
# ── From Python lists ──────────────────────────────────────────────────────────
a1d = np.array([10, 20, 30, 40, 50])
a2d = np.array([[1, 2, 3],
                [4, 5, 6]])
a3d = np.array([[[1,2],[3,4]],
                [[5,6],[7,8]]])

print("1-D:", a1d)
print("2-D:\n", a2d)
print("3-D shape:", a3d.shape)
print(a1d.ndim)
print(a2d.ndim)
print(a3d.ndim)
```

```python
# ── Zeros, Ones, Full, Eye, Identity ──────────────────────────────────────────
print("zeros(3,4):\n",    np.zeros((3, 4)))
print("ones(2,5):\n",     np.ones((2, 5)))
print("full(3,3, 7):\n",  np.full((3, 3), fill_value=7))
print("eye(4):\n",        np.eye(4))
print("identity(3):\n",   np.identity(3))
print("empty(2,3):\n",    np.empty((2, 3)))      # uninitialized — fast
```

```python
# ── Ranges & Sequences ────────────────────────────────────────────────────────

# arange — like Python range but returns ndarray
print("arange(0,20,2):", np.arange(0, 20, 2))

# linspace — N evenly-spaced points between start & stop (inclusive)
print("linspace(0,1,6):", np.linspace(0, 1, 6))

# logspace — evenly spaced on a log scale (great for hyperparameter grids)
print("logspace(0,3,4):", np.logspace(0, 3, 4, base= 10))   # 10^0 … 10^3. Default Base value= 10

# geomspace — geometric sequence
print("geomspace(1,1000,4):", np.geomspace(1, 1000, 4))

# meshgrid — build coordinate grids (used in contour/surface plots)
x = np.linspace(-2, 2, 5)
print("X: ",x)
y = np.linspace(-2, 2, 5)
print("Y: ",y)
XX, YY = np.meshgrid(x, y)
print("meshgrid XX:\n", XX)
print(YY)
```

```python
arr = np.random.randn(1000, 50).astype(np.float32)
print(arr)
print(f"ndim        : {arr.ndim}")           # number of dimensions
print(f"shape       : {arr.shape}")          # (rows, cols)
print(f"size        : {arr.size:,}")         # total elements
print(f"dtype       : {arr.dtype}")          # data type
print(f"itemsize    : {arr.itemsize} bytes") # bytes per element
print(f"nbytes      : {arr.nbytes/1024:.1f} KB")  # total memory
```

```python
# Sales dataset: rows=employees, cols=[Q1,Q2,Q3,Q4,Annual_Target]
np.random.seed(0)
sales = np.random.randint(50_000, 200_000, (8, 5))
print("Sales matrix (8 employees × 5 columns):\n", sales)
```

```python
# ── Basic indexing ────────────────────────────────────────────────────────────
print("Employee 0, all quarters  :", sales[0])
print("Employee 2, Q3 (col 2)    :", sales[2, 2])
print("Last employee             :", sales[-1])
print("Last 2 employees, Q1-Q2   :\n", sales[-2:, :2])
print("Every other employee      :\n", sales[::2])
print("Reversed employees        :\n", sales[::-1])
```
